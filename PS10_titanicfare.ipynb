```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df = sn.load_dataset('titanic')

df.head()
```

```python
plt.hist(df['fare'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Titanic Fares')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python
# how the price of the ticket for each passenger is distributed using distplot()
sns.distplot(df['fare'])
```

```python
sns.distplot(df['fare'], kde=False)
```

```python
sns.distplot(df['fare'], kde=False, bins=5)
```
