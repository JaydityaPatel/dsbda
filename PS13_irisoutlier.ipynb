```python
col = 'SepalWidthCm'

# Calculate IQR
Q1 = data1[col].quantile(0.25)
Q3 = data1[col].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Get outliers
outliers = data1[(data1[col] < lower_bound) | (data1[col] > upper_bound)]

# Print info
print(f"Outliers in column '{col}':")
print(outliers)
```
