```python
def graph(y):
    sns.boxplot(x='species',y=y , data=data1)
    plt.title(f"Boxplot of {y} by species")
plt.figure(figsize=(10,10))
plt.subplot(221)
graph('SepalLengthCm')
plt.subplot(222)
graph('SepalWidthCm')
plt.subplot(223)
graph('PetalLengthCm')
plt.subplot(224)
graph('PetalWidthCm')
plt.tight_layout()
plt.show()
```
